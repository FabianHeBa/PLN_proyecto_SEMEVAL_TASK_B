```python
import pandas as pd
```

```python
path = os.path.join('..', 'data', 'task_b_training_set.parquet')

training = pd.read_parquet(path)
```

```python
languages = training['language'].value_counts()

languages
```

```python
labels = [1,7,10]
llm_python = training[training['label'].isin(labels)]

llm_python = llm_python[llm_python['language']=='Python']

n_samples_python = len(llm_python)
```

```python
n_samples_python
```

```python
llm_python['label'].unique()
```

```python
human = training[training['label']==0]

human_python = human[human['language']=='Python']

stratified_hp = human_python.sample(n_samples_python,random_state=17)

```

```python
llm_java = training[training['label'].isin(labels)]

llm_java = llm_java[llm_java['language']=='Java']

n_samples_java = len(llm_java)
```

```python
n_samples_java
```

```python
human_java = human[human['language']=='Java']

stratified_hj = human_java.sample(n_samples_java,random_state=17)

len(stratified_hj)
```

```python
llm_c_sharp = training[training['label'].isin(labels)]

llm_c_sharp = llm_c_sharp[llm_c_sharp['language']=='C#']

n_samples_c_sharp = len(llm_c_sharp)
```

```python
llm_c_sharp['label'].unique()
```

```python
human_c_sharp = human[human['language']=='C#']

stratified_hcs = human_c_sharp.sample(n_samples_c_sharp,random_state=17)

len(stratified_hcs)
```

```python
human_total = [stratified_hp,stratified_hj,stratified_hcs]
human_concat = pd.concat(human_total)

llm_total = [llm_python,llm_java,llm_c_sharp]
llm_concat = pd.concat(llm_total)

stratified = pd.concat([human_concat,llm_concat])

file = 'stratified_sample.csv'

stratified.to_csv(file,index=False)
```
